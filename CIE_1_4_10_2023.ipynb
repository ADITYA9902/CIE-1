```python
import pandas as pd
data={'productname'   :['mobile','tshirts','detergent','tv','pant'],
      'category'      :['electronics','clothing','home & kitchen','elkectronics','clothing'],
      'price'        :[5000,150,50,10000,300],
      'quantity_sold':[2,3,5,2,3],
}
df=pd.DataFrame(data)
df['total_revenue']=df['price']*df['quantity_sold']
pd.pivot_table(df,index='category',values=['total_revenue','price'],aggfunc={'total_revenue':'sum','price':'mean'})
```
